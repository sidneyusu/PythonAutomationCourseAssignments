**PROBLEM 1:**
Write a Python program which accepts the radius of a circle from the user and compute the area.

Algorithm:
- Start
- Import pi variable from the math package
- Get user input and assign to variable radius
- Calculate the area of the radius and assign to variable area
- Display area
- Stop

Flowchart:

```python
from math import pi

radius = float(input("Input the radius of the circle: "))
area = pi * radius**2

print('The area of a circle with a radius of ' + str(radius) + ' is ' + str(area))
```

**PROBLEM 3:**
Write a Python program to get the difference between a given number and 17, if the number is greater than 17 return double the absolute difference.

Algorithm:
- Start
- Declare function named absoluteDiff with parameter num
- Declare variables num and dif
- Get the difference between num and 17 and assign the result to variable dif
- If dif is greater than 17 then multiply the absolute of dif with 2 and return the result
- Otherwise return dif
- Stop

Flowchart:

```python
def absoluteDiff(num):
    dif = num - 17
    if dif > 17:
        return abs(dif)*2
    else:
        return dif
    
num = 36
print(absoluteDiff(num))
```

**PROBLEM 4:**
Write a Python program to test whether a number is within 100 of 1000 or 2000.

Algorithm:
- Start
- Declare variable num
- Read value for num
- If num is less than 100 display statement that the number is within 100
- Else if num is less than 1000 display statement that the number is within 1000
- Else if num is less than 2000 display statement that the number is within 2000
- Otherwise do nothing
- Stop

Flowchart:

```python
num = 100
if num < 100:
    print(str(num) + ' is within 100')
elif num < 1000:
    print(str(num) + ' is within 1000')
elif num < 2000:
    print(str(num) + ' is within 2000')
```

**PROBLEM 5:**
Write a Python program to calculate the sum of three given numbers, if the values are equal then return three times of their sum.

Algorithm:
- Start
- Declare function named sumOfThree with parameters num1, num2, num3
- Declare variable sum
- Add num1, num2, and num3 and assign the result to a variable sum
- If num1, num2, and num3 are equal values then multiple sum with 3 and return the result
- Otherwise return sum
- Stop

Flowchart:

```python
def sumOfThree(num1, num2, num3):    
    sum = num1 + num2 + num3
    if num1 == num2 and num2 == num3:        
        return sum * 3
    else:
        return sum

num1 = 2
num2 = 3
num3 = 5

print('The result is ' + str(sumOfThree(num1, num2, num3)))
```
